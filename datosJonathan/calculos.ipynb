```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_exp_Nor, y_exp_Nor = np.loadtxt(fname="Normalizado.csv", delimiter=';', unpack=True)
x_exp_med, y_exp_med = np.loadtxt(fname="Media.csv", delimiter=';', unpack=True)
```

```python
plt.title("Campo Magnitud")

plt.plot(x_exp_Nor, y_exp_Nor, '.', label = "Exp. data")
plt.plot(x_exp_med, y_exp_med, '.', label = "media")
plt.legend()
plt.grid()
plt.xlabel("Posicion y(cm)")
plt.ylabel("Campo magnetico")
```
